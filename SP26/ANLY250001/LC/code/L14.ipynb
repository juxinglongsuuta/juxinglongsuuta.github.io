```python
# for loop to collect a few grades, then calculate the average grade
# define accumulator/counter variables
sumG=0.0 # sum of the grades
nStu=0

print("Input the number of students:")
nStr=input()
nStu=int(nStr)


# for loop to collect a few grade from keyboard
for i in range(nStu):
  print(f"\nInput student {i+1} grade:")
  gStr=input()
  g=float(gStr)

  # deposit into a bank account
  sumG=sumG+g

# calculate the average
a = sumG/nStu
print(f"\nNumber of students: {nStu}")
print(f"Average grade: {a:.2f}")
```

```python
# math addition quiz (for loop)

import random # use the code in the random file

# define vars
n=0# total number of questions
i=0 # the current number of questions already shown
nc=0 # number of correct answers
nw=0 # number of incorrect/wrong answers
cp=0 # currect points/score

# ask how many questions
print("Input the number questions to generate:")
nstr=input()
n=int(nstr)

# for loop for each question
print("\n")
for c in range(n):
  # gen one question
  i=i+1
  n1=random.randint(0,9)
  n2=random.randint(0,9)
  gt=n1+n2 # gt is the ground truth
  s=f"\nQ{i}: {n1} + {n2} = ?"
  print(s)
  astr=input()
  a=int(astr) # user answer
  if a==gt:
    print("Correct!")
    nc=nc+1
    cp=cp+2
  elif a!=gt:#else:
    print("Incorrect!")
    nw=nw+1
    cp=cp-3


# summary information
print("\n\nSummary Information")

s=f"Total number of questions: {n}"
print(s)

r=nc/n
s=f"Correct answers: {nc} / {n} ( {r:.2%} )"
print(s)

r=nw/n
s=f"Incorrect answers: {nw} / {n} ( {r:.2%} )"
print(s)

s=f"Final points: {cp:.2f}"
print(s)

if cp>0:
  print("Congrats!")
elif cp==0:
  print("Not bad.")
elif cp<0: #else:
  print("You need to work hard!")
```
