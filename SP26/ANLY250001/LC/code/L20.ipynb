```python
import pandas as pd
mdf=pd.read_csv("uscity.csv")
mdf # print(mdf)
# mdf: dataframe, table
```

```python
df10=mdf.head(10) # new dataframe, table
df10
```

```python
# filtering: result in table
# dataframe, table vs single row, single column, series
# and: &
# or: |
# not: ~
c=df10["state"]=='Texas' # series
print(c)
ndf=df10[c] # table, dataframe
ndf
```

```python
#
c1=df10["state"]=="New York"
print(c1)

c2=df10["state"]=="Texas"
print(c2)

c3=df10["state"]=="Florida"
print(c3)
# & |
# &
c=c1|c2|c3 # &
print(c)
df10[c]
```

```python
#
c1=mdf["state"]=="Texas"
c2=mdf['population']>1000000
c3=mdf['population']<6000000
c=c1&c2&c3
mdf[c]#
```

```python
# Texas or Florida
c1=mdf["state"]=="Texas"
c4=mdf["state"]=="Florida"
c2=mdf['population']>1000000
c3=mdf['population']<6000000
c=(c1|c4) & c2 & c3 #
df=mdf[c]# table
df.to_csv("texas_florida_city.csv",index=False)
df
```
