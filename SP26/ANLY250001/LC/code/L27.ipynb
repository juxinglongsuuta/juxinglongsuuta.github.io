```python
# max() min() avg() sum() count()
# fun will work on a signle table, result in a single value
%%sql
select max(Population) as MaxP,
       min(Population) as MinP,
       sum(Population) as TotalP

from data --
;
```

```python
%%sql
select max(Population) as MaxP,
       min(Population) as MinP,
       sum(Population) as TotalP,
       count(*) as N

from data --
;
```

```python
## group by: divide a single table into multiple tables
## select, from, where, group by, having
%%sql
select max(Population) as MaxP,
       min(Population) as MinP,
       sum(Population) as TotalP,
       count(*) as N,
       State
from data --
group by State -- 51 tables
order by count(*) desc --N desc
;
```

```python
# having after group by
# having: filtering, filter tables, must aggregation funs
# where: filtering, filter rows, no aggregation funs
%%sql
select max(Population) as MaxP,
       min(Population) as MinP,
       sum(Population) as TotalP,
       count(*) as N,
       State
from data --
group by State -- 51 tables
having count(*)>20
order by count(*) desc --N desc
;
```
