```python
#
print("Input the name:")
nStr=input()

print("Input the grade:")
gStr=input()
gDbl=float(gStr)

# A B C D F
lgStr=""
if gDbl>=90:
  print("A")
  lgStr="A"
elif gDbl>=80:
  print("B")
  lgStr="B"
elif gDbl>=70:
  print("C")
  lgStr="C"
elif gDbl>=60:
  print("D")
  lgStr="D"
else:
  print("F")
  lgStr="F"

#
infStr=f"Name: {nStr}, Letter Grade: {lgStr}"
print(infStr)
```

```python
# Weight loss practice

# define variables (important vars: counter/accum vars)
name=""
sw=0.0 # start weight
p=0.0 # loss percent each month
gw=0.0 # goal weight

cw=0.0 # current weight at any time
n=0 #current month

# get input
print("Name:")
name=input()# string

print("Start weight: (lbs)")
swstr=input()
sw=float(swstr)

print("Percentage:")
pstr=input()
p=float(pstr)

print("Goal weight: (lbs)")
gwstr=input()
gw=float(gwstr)

#  print out some infor
infor=f"Name: {name}"
print(infor)

infor=f"Start weight: {sw:.3f} lbs"
print(infor)

infor=f"Percentage: {p:.2%}"
print(infor)


#month 0
cw=sw

# month 1
n=n+1
loss=cw*p
cw=cw-loss

infor=f"After month {n}, lose {loss} lbs"
print(infor)

infor=f"After month {n}, current weight: {cw} lbs"
print(infor)


# month 2
n=n+1
loss=cw*p
cw=cw-loss

infor=f"After month {n}, lose {loss} lbs"
print(infor)

infor=f"After month {n}, current weight: {cw} lbs"
print(infor)

# month 3
n=n+1
loss=cw*p
cw=cw-loss

infor=f"After month {n}, lose {loss} lbs"
print(infor)

infor=f"After month {n}, current weight: {cw} lbs"
print(infor)

# summary infor

```
