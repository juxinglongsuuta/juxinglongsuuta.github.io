```python
i = 1
while i>0: # if i > 0
  # loop body
  print("\nLoop body repeaded one more time")
  print(f"{i} > 0")
  print("Input an integer:")
  iStr=input()
  i=int(iStr)
  print(f"i = {i}\n")

print("This is the next line of the loop")
print(f"i = {i}")
```

```python
# collection (warehouse)
c1=[2,3,6]
c2=range(20) # 20 items in warehouse, 0,...,19
```

```python
for c in c1:# step 1: c=2 step 2: c=3 step 3: c=6
  print(c)
  print(c)

print("The next of loop")
```

```python
for c in range(20):# step 1: c=0 step 2: c=1
  print(c)
  print(c)

print("The next of loop")
```

```python
# break continue
# break: stop the loop
# continue: stop the current step, process next item
for c in range(20):# step 1: c=0 step 2: c=1
  print(c)
  break
  print(c)

print("The next of loop")
```

```python
# break continue
# break: stop the loop
# continue: stop the current step, process next item
for c in range(20):# step 1: c=0 step 2: c=1
  print(c)
  if c==3:
    break
  print(c)

print("The next of loop")
```

```python
# break continue
# break: stop the loop
# continue: stop the current step, process next item
for c in range(5):# step 1: c=0 step 2: c=1
  print(c)
  if c==3:
    continue
  print(c)

print("The next of loop")
```

```python
# infinite loop: break
while True:
  print("\nLoop body")
  print("Do you want to continue the loop? (y/n)")
  astr=input()
  if astr=="n":
    break

print("Next line of loop")
```

```python
# number guessing game
fp=0 # final points
import random
gt=random.randint(1,10) # ground truth

while True:
  print("Guess a number:")
  nstr=input()
  n=int(nstr) # use guess number
  if n==gt:
    fp=fp+5
    print("Congrats! You gain 5 points!")
    break# stop the game
  else:
    fp=fp-1
    print("Sorry! You lose 1 point!")

print(f"Final points: {fp}")
```
