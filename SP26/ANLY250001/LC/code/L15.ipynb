```python
# while loop to collect a few grades
# then calculate the average grade
# define variables
nStu=0 # number of stduents
sumG=0 # sum of the grades

while True: # infinite loop
  # loop body
  nStu=nStu+1
  print(f"\nInput student {nStu} grade:")
  gStr=input()
  g=float(gStr)

  sumG=sumG+g

  # another student grade
  print(f"\nAnother student grade? (y/n)")
  userChoice=input()
  if userChoice=='n':
    break # stop the loop


a = sumG/nStu
print(f"\nNumber of students: {nStu}")
print(f"Average grade: {a:.2f}")

```

```python
# Student grades using a while loop
# Quiz 2: use a while loop

# define vars
n=0 # number of students
na=0 # number of A's
nb=0
nc=0
nd=0
nf=0

sg=0 # the sum of all grades
sga=0 # sum of A grades
sgb=0
sgc=0
sgd=0
sgf=0

# loop
# for loop: the repeated times
# while loop: do not know how many time to repeat

# use a while loop to collect students' grade
# infinite loop
while True:
  n=n+1
  print(f"\nInput student {n} grade:")
  gstr=input()# string
  g=float(gstr)
  sg=sg+g

  # determine the letter grade
  if g>=90:
    lg="A"
    na=na+1
    sga=sga+g
  elif g>=80:
    lg="B"
    nb=nb+1
    sgb=sgb+g
  elif g>=70:
    lg="C"
    nc=nc+1
    sgc=sgc+g
  elif g>=60:
    lg="D"
    nd=nd+1
    sgd=sgd+g
  elif g<60:#else:
    lg="F"
    nf=nf+1
    sgf=sgf+g

  print(f"The letter grade: {lg}")

  print("\nIs there another student's grade?(y/n)")
  astr=input()
  if astr=="n":
    break # stop the loop

# summary
print("\n\nSummary Information")

print(f"\nNumber of students: {n}")

a=sg/n
print(f"\nAverage grade: {a:.2f}")
a=sga/na
print(f"Average A grade: {a:.2f}")
a=sgb/nb
print(f"Average B grade: {a:.2f}")
a=sgc/nc
print(f"Average C grade: {a:.2f}")
a=sgd/nd
print(f"Average D grade: {a:.2f}")
a=sgf/nf
print(f"Average F grade: {a:.2f}")

print("\n")
r=na/n
print(f"A: {na}/{n} ({r:.2%})")
r=nb/n
print(f"B: {nb}/{n} ({r:.2%})")
r=nc/n
print(f"C: {nc}/{n} ({r:.2%})")
r=nd/n
print(f"D: {nd}/{n} ({r:.2%})")
r=nf/n
print(f"F: {nf}/{n} ({r:.2%})")
```
