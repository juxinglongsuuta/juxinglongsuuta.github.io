```python
# Sales Analysis using a while loop

# define vars
n=0 # number of orders
ss=0 # sum of subtotals/total sales/total revenue
avs=0 # average subtotal of all orders
tnc=0 # total number of cogs sold
tng=0 # total number of gears sold
ssc=0 # sum of cogs sales
ssg=0 # sum of gears sales
nl1=0 # number of orders in level 1: subtotal<1000
nl2=0 # level 2: 1000<=subtotal<=2000
nl3=0 # level 3: subtotal>2000

# collect orders using a while loop
while True:
  # collect one order
  n=n+1

  print(f"\nN cogs in Order {n}:")
  cstr=input()
  c=float(cstr) # c is number of cogs sold in one order

  print(f"N gears in Order {n}:")
  gstr=input()
  g=float(gstr) # g is number of gears in one order

  tnc=tnc+c
  tng=tng+g

  # 60/each cogs  200/each gears
  ssc=ssc+c*60
  ssg=ssg+g*200

  s=c*60+g*200 # s is the subtotal for one order
  ss=ss+s

  if s<1000:
    print(f"Subtotal: ${s:.2f}, below $1000")
    nl1=nl1+1
  elif s<=2000:
    print(f"Subtotal: ${s:.2f}, between $1K and $2K")
    nl2=nl2+1
  elif s>2000:
    print(f"Subtotal: ${s:.2f}, above $2000")
    nl3=nl3+1

  print("\nAnother order?(y/n)")
  astr=input()# answer string by user
  if astr=="n":
    break # stop the loop

# summary information
print("\n\nSummary Information")

print(f"N orders: {n}")

print(f"Sum of subtotal: ${ss:.2f}")
print(f"Sum of cogs sale: ${ssc:.2f}")
print(f"Sum of gears sale: ${ssg:.2f}")

a=ss/n
print(f"Average subtotal: ${a:.2f}")

print(f"Total N cogs sold: {tnc}")
print(f"Total N gears sold: {tng}")

r=nl1/n
print(f"L1: {nl1}/{n}, {r:.2%}")
r=nl2/n
print(f"L2: {nl2}/{n}, {r:.2%}")
r=nl3/n
print(f"L3: {nl3}/{n}, {r:.2%}")
```
