```python
%%sql
select City as C,
       State as S,
       Population as P,
       Population+1000 as NP1,
       Population*1.03 as NP2
from data
limit 6
;
-- sql: result in a table
```

```python
%%sql
select *
from data
where State='Utah' or State='Texas' -- filtering
limit 10
;
```

```python
%%sql
select *
from data
where Population>1000000 and State='Texas'
order by Population desc -- asc desc
;
```

```python
%%sql
select *
from data
where State='Utah'
order by Population desc
limit 3
;
```

```python
# avg() max() min() sum() count()
%%sql
select sum(Population),avg(Population),State
from data -- result in a single table
group by State
;
-- fun: will work on a single, result in a single row
-- group by: divide a single table into small tables
```
