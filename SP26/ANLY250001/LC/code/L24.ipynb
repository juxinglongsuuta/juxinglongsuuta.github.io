```python
import pandas as pd
tips_df=pd.read_csv("tips.csv")
tips_df
```

```python
# filtering:
# > >= <= ==
# series -> condition
# and: &
# or: |
# groupby
# 9
c1=tips_df['sex']=='Female'
c2=tips_df['total_bill']>20
c3=tips_df['tip']>5
c=c1&c2&c3
df=tips_df[c]
ndf=df.sort_values(by=["total_bill","tip"],ascending=[False,False])
ndf.to_csv("female_filter_tips.csv",index=False)
ndf
# create a new table (dataframe)
```

```python
%ls
```

```python
ndf
```

```python
s1=ndf['tip']
n=s1.max()
print(f"The max tip is: ${n:.2f}")
```

```python
# 10
c1=tips_df['tip']>=6
c2=tips_df['tip']<=8
c=c1&c2
ndf=tips_df[c]
ndf
```

```python
s1=ndf['total_bill']
n=s1.max()
print(f"The max total bill is: ${n:.2f}")
```

```python
# 11
s1=tips_df['tip']
s2=tips_df['total_bill']
tips_df["tip_ratio"]=s1/s2
tips_df
```
