```python
import pandas as pd
tips_df=pd.read_csv("tips.csv")
tips_df
```

```python
# 7
# groupby(): work with/on table/dataframe,
# result in multiple tables/dfs
# divide a single table into small tables
mdf=tips_df
mdf
```

```python
mdf.columns
```

```python
s1=mdf['tip'] # [] .
s2=mdf['total_bill']
mdf["ratio"]=s1/s2
mdf
```

```python
vs=mdf['day'].unique() # series
print(vs)
n=len(vs)
print(n)
```

```python
# groupby():
r=mdf.groupby(by=['day'])
n=len(r)
```

```python
r.groups.keys()
```

```python
r.get_group("Fri")# dataframe/table
```

```python
%ls
```

```python
print(vs)
for v in vs: # v: placeholder/container v='Sun' v='Sat'
  # loop body
  print(v)
  print(f"\n\n\nThis data is for day: {v}")
  sdf=r.get_group(v)
  ndf=sdf.sort_values(by=['tip'],ascending=[False])
  ndf.to_csv(f"{v}_tips.csv",index=False)
  n=len(ndf)
  print(n)
  print(f"N orders on {v}: {n}")
  s=ndf['total_bill']
  # unique() max() min() mean() sum() work with series
  n=s.sum() #
  print(f"Sum of total bills for {v}: ${n:.2f}")
  n=s.mean()
  print(f"Average total bill for {v}: ${n:.2f}")
  n=s.max()
  s=ndf['size']
  n=s.sum()
  print(f"For {v}, there are {n} customers.")
  s=ndf['tip']
  n=s.sum()
  print(f"Sum of tips on {v}: ${n:.2f}")
  n=s.mean()
  print(f"Average tip on {v}: ${n:.2f}")

  print(sdf)
  print(ndf)
```
