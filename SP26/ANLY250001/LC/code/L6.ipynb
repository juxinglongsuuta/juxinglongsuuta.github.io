```python
# if else statement
# your name and student id here
print("Input your current salary: ")
salStr=input() # string

salDbl=float(salStr) # convert into float/double

nSalDbl= salDbl*1.1 # salary increased by 10%

if salDbl>100000:
  print("The current salary is greater than 100K")
else:
  print("The current salary is not greater than 100K")

inforStr=f"The new salary is: ${nSalDbl:.2f}"# in RAM
print(inforStr)
```

```python
#
print("Input a grade:")
gstr=input()#
# float(double) 8B / int(4B)
g=float(gstr)

if g>=80:
  # code block 1
  print("B or above")
  print("Congrats!")
elif g>=70:
  # code block 2
  print("C")
elif g>=60:
  # code block 3
  print("D")
elif g>=50:
  # code block 4
  print("E")
else:
  print("NOT PASS")

print("After if else. We still continue")
```
