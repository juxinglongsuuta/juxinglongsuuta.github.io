```python
import pandas as pd
```

```python
tips_df=pd.read_csv("tips.csv")
```

```python
tips_df
```

```python
tips_df.columns
```

```python
# 6
r=tips_df.groupby(by=['sex']) #

# s=tips_df['sex']
# vs=s.unique()

vs=r.groups.keys()

for i in vs:
  print(f"\n\nOrders by: {i}")
  sdf=r.get_group(i)

  #
  n=len(sdf)
  print(f"Number of orders: {n}")

  #
  s=sdf['size']
  n=s.sum()
  print(f"Number of customers served: {n}")

  #
  t=sdf['total_bill'].sum()
  print(f"Sum of total bill: $ {t:.2f}")

  t=sdf['tip'].sum()
  print(f"Sum of tips: {t}")

  # average tip
  t=sdf['tip'].mean()
  print(f"Average tip: {t}")

  sdf.to_csv(f"{i}_tips.csv",index=False)
  print(sdf)
# print(vs)
```
