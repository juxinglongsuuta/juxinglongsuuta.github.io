```python
%%sql
select *
from sale
limit 10
;
```

```python
%%sql
select sum(SalePrice*Quantity) as Revenue,
       Category
from sale
group by Category
order by Revenue desc
;
```

```python
# count() max(): aggregation
#
# 3
%%sql
select count(*) as NRows
from sale
;
```

```python
%%sql
select count(*) as NRows,
       Category
from sale
group by Category
;
```

```python
# where: No aggregation function
# having: Must use aggregation function
%%sql
select count(*) as NRows,
       Category
from sale
group by Category
having count(*)>500
;
```

```python
%%sql
select count(*) as NRows,
       Category
from sale
group by Category
having count(*)>500
order by NRows desc
;
```

```python
%%sql
select count(*) as NRows,
       Category
from sale
group by Category
having count(*)>500
order by count(*) desc
;
```

```python
# 4
%%sql
select sum(SalePrice*Quantity) as TotalRevenue
from sale
;
```

```python
%%sql
select sum(SalePrice*Quantity) as TotalRevenue,
       Category
from sale
group by Category
order by TotalRevenue desc
;
```

```python
%%sql
select sum(SalePrice*Quantity) as TotalRevenue,
       Category
from sale
group by Category
order by sum(SalePrice*Quantity) desc
;
```

```python
%%sql
select sum(SalePrice*Quantity) as TotalRevenue,
       Category
from sale
group by Category
having sum(SalePrice*Quantity)>30000
order by sum(SalePrice*Quantity) desc
;
```

```python
# 5
%%sql
select sum(Quantity) as NPairsSold,
       Category
from sale
group by Category
;
```

```python
%%sql
select sum(Quantity) as NPairsSold,
       Category
from sale
group by Category
having sum(Quantity)<500
order by NPairsSold asc
;
```
