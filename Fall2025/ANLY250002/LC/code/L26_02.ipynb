```python
%%sql
select *
from sale
limit 10
;
```

```python
%%sql
select sum(SalePrice*Quantity) as Amt,
       OrderID
from sale
group by OrderID
limit 10
;
```

```python
# sql: table
# select: columns
# from: a single table
# where: filtering (delete rows) a single table
# group by:
#    divide a single tables into multiple tables
# having: filtering tables (delete tables)
# order by:
# max() min() avg() count() sum()
#    : work on single table, result in a single row
%%sql
select 	SalePrice,	Quantity,
        SalePrice*Quantity as Subtotal
from sale
limit 10
;
```

```python
%%sql
select 	SalePrice*Quantity as Subtotal
from sale
limit 10
;
```

```python
%%sql
select sum(SalePrice*Quantity) as Revenue
from sale
limit 10
;
```

```python
%%sql
select sum(SalePrice*Quantity) as Revenue,
       SaleDate
from sale
group by SaleDate
limit 10
;
```

```python
%%sql
select sum(SalePrice*Quantity) as Revenue,
       SaleDate
from sale
group by SaleDate
order by Revenue desc
limit 10
;
```

```python
%%sql
select sum(SalePrice*Quantity) as Revenue,
       CustomerID,
       FirstName
from sale
group by CustomerID
order by Revenue desc
limit 10
;
```

```python
%%sql
select sum(Quantity) as NSold
from sale
limit 10
;
```

```python
%%sql
select sum(Quantity) as NSold,
       ProductName
from sale
group by ProductName
order by NSold asc
limit 10
;
```

```python
%%sql
select sum(Quantity) as NSold,
       ProductName
from sale
group by ProductName
having sum(Quantity)>100
order by NSold desc
limit 10
;
```
