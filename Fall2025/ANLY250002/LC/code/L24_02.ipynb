```python
%%sql
select City as C,
       State,
       Population,
       Population as P,
       Population+100 as P100
from data
limit 3
;
```

```python
%%sql
-- > < <= >= <> equal to =
-- and or
select City as C,
       State,
       Population,
       Population as P,
       Population+100 as P100
from data
where State='Utah'
;
```

```python
%%sql
select City as C,
       State,
       Population,
       Population as P,
       Population+100 as P100
from data
where State='Utah' or
      State='Nevada' or
      State='Ohio'
;
```

```python
%%sql
select City as C,
       State,
       Population,
       Population as P,
       Population+100 as P100
from data
where Population>1000000
;
```

```python
%%sql
select City as C,
       State,
       Population,
       Population as P,
       Population+100 as P100
from data
where Population>1000000 and
      Population<2000000
;
```

```python
%%sql
select City as C,
       State,
       Population,
       Population as P,
       Population+100 as P100
from data
where Population>1000000 and
      Population<2000000 and
      State='Texas'
;
```

```python
%%sql
-- max() min() avg() count() sum()
select count(*) as N,
       max(Population) as MAX,
       min(Population) as MIN
from data
;
```

```python
%%sql
-- group by: divide a single table into multiple tables
select count(*) as N,
       max(Population) as MAX,
       min(Population) as MIN,
       State
from data
group by State
;
```

```python
%%sql
-- order by: asc desc
select count(*) as N,
       max(Population) as MAX,
       min(Population) as MIN,
       State
from data
group by State
order by N desc
;
```

```python
%%sql
-- having
select count(*) as N,
       max(Population) as MAX,
       min(Population) as MIN,
       State
from data
group by State
having count(*)>30
order by N desc
;
```
