```python
# while loop to collect a few grades
# then calculate the average grade
# define variables
nStu=0 # number of stduents
sumG=0 # sum of the grades

while True: # infinite loop
  # loop body
  nStu=nStu+1
  print(f"\nInput student {nStu} grade:")
  gStr=input()
  g=float(gStr)

  sumG=sumG+g

  # another student grade
  print(f"\nAnother student grade? (y/n)")
  userChoice=input()
  if userChoice=='n':
    break # stop the loop


a = sumG/nStu
print(f"\nNumber of students: {nStu}")
print(f"Average grade: {a:.2f}")

```

```python
# sales analysis
#
# define variables (counter/accumulator)
n=0 # number of orders
nc=0 # total number of cogs sold
ng=0 # total number of gears sold

nl1=0 # number of orders with subtotal <1000
nl2=0 # between 1000-2000
nl3=0

sums=0 # total revenue/ sum of all subtotals
sumc=0 # sum of cog revenue
sumg=0 # total revenue for gears

# while loop to collect orders
while True:
  # one more order
  n=n+1 #
  print(f"\nNumber of cogs in Order {n}:")
  cstr=input()
  c=int(cstr)
  nc=nc+c #
  print(f"Number of gears in Order {n}:")
  gstr=input()
  g=int(gstr)
  ng=ng+g #

  s=c*60 + g*200 # subtotal
  sums=sums+s #
  sumc=sumc+c*60 #
  sumg=sumg+g*200 #

  if s<1000:
    print(f"Subtotal: ${s:.2f}, below $1000")
    nl1=nl1+1#
  elif s<=2000:
    print(f"Subtotal: ${s:.2f}, between $1000-$2000")
    nl2=nl2+1 #
  elif s>2000:#else:
    print(f"Subtotal: ${s:.2f}, above $2000")
    nl3=nl3+1 #

  print(f"\nAnother order? (y/n)")
  userchoice=input()
  if userchoice=="n":
    break# stop the loop


# summary information
print(f"\n\nSummary Information")
print(f"N orders: {n}")
print(f"N cogs: {nc}")
print(f"N gears: {ng}")

print(f"\nTotal revenue: ${sums:.2f}")
print(f"Revenue for cogs: ${sumc:.2f}")
print(f"Revenue for gears: ${sumg:.2f}")

print(f"Average subtotal: ${sums/n:.2f}")

print(f"\nL1: {nl1}/{n}, {nl1/n:.2%}")
print(f"L2: {nl2}/{n}, {nl2/n:.2%}")
print(f"L3: {nl3}/{n}, {nl3/n:.2%}")
```
