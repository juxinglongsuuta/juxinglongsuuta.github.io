```python
# 1
import pandas as pd
tips_df=pd.read_csv("tips.csv")#
```

```python
# 2
cols=tips_df.columns #
print(cols)

n=len(cols) # RAM
print(f"N columns: {n}")
```

```python
# 3
tips_sorted_df=tips_df.sort_values(by=['total_bill', 'tip'],ascending=[False,False])
tips_sorted_df.to_csv("tips_sorted.csv",index=False)
```

```python
# 4
s=tips_df['day']
u=s.unique()
print(u)

n=len(u) #
print(f"N of unique values in day column: {n}")

c=s.value_counts()
print(c)
```

```python
s=tips_df['time']
u=s.unique()
print(u)

n=len(u) #
print(f"N of unique values in time column: {n}")

c=s.value_counts()
print(c)
```

```python
# 5
s=tips_df['total_bill'] # series
t=s.sum() #
print(f"Total revenue: $ {t:.3f}")

n=len(s)
n=s.count()
print(f"N of values: {n}")

m=s.mean()
print(f"Average bill: ${m:.2f}")

m=s.max()
print(f"Largest bill: {m}")
```
