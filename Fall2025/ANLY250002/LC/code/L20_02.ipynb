```python
# groupby() and for loop
# groupby() will work on table
# divide a table into small ones (tables)
```

```python
r=df1.groupby(by=["State"])# 51 in r
states=["Nevada",'Utah',"Texas"]
n=len(states)# RAM
print(f"There are {n} states.")

for i in states:# i="Nevada" i="Utah" i="Texas"
  print("\n\nOne iteration")
  print(f"The current state is: {i}")
  sdf=r.get_group(i)# table RAM
  sdf=sdf.sort_values(by=["Population"],ascending=[False])
  #
  sdf.to_csv(f"d1/{i}.csv",index=False)
  print(sdf.head(3))
```
