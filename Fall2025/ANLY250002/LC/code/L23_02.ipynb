```python
%%sql
select *
from data
limit 6
;
```

```python
%%sql
select City as C,
       State as S,
       Population as P
from data
limit 6
;
```

```python
%%sql
select City as C,
       State as S,
       Population as P,
       Population+1000 as P1000,
       Population*1.03 as P3More
from data
limit 6
;
```

```python
%%sql
select City as C,
       State as S,
       Population as P,
       Population+1000 as P1000,
       Population*1.03 as P3More
from data
where State='Utah'
;
```

```python
%%sql
select City as C,
       State as S,
       Population as P,
       Population+1000 as P1000,
       Population*1.03 as P3More
from data
where State='Utah' and
      Population>100000
;
# and or
```

```python
%%sql
select City as C,
       State as S,
       Population as P,
       Population+1000 as P1000,
       Population*1.03 as P3More
from data
where State='Utah' or
      State='Nevada'
;
```

```python
%%sql
select City as C,
       State as S,
       Population as P,
       Population+1000 as P1000,
       Population*1.03 as P3More
from data
where State='Utah'
order by Population desc
;
# asc desc
```

```python
%%sql
select count(*) as N,
       min(Population) as M,
       City
from data
where State='Utah'
;
# max() min() avg() count()
```

```python
%%sql
select count(*) as N,
       State
from data
group by State
;
```
