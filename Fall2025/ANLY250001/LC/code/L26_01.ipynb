```python
# subtotal for each row
# sql: result in a single table
# select: columns
# from: result in a single table
# where: filtering (delete row), in a single table
# group by: divide a single table in multiple
#         : only way that will result in multiple tabs
# having: filtering multiples
# order by:
# max() min() count() sum() avg()
#     : work on a single table, result in a single row
%%sql
select SaleDate,
       CustomerID,
       FirstName, LastName,
       OrderID,
       ProductName,
       Category,
       Color,
       SalePrice*Quantity as Subtotal,
       ManufacturerName
from sale
limit 10
;
```

```python
%%sql
select SaleDate,
       CustomerID,
       FirstName, LastName,
       OrderID,
       ProductName,
       Category,
       Color,
       SalePrice*Quantity as Subtotal,
       sum(SalePrice*Quantity) as Revenue,
       ManufacturerName
from sale
;
```

```python
%%sql
select sum(SalePrice*Quantity) as Revenue,
       SaleDate
from sale
group by SaleDate
limit 10
;
```

```python
%%sql
select sum(SalePrice*Quantity) as Revenue,
       CustomerID,
       FirstName,
       LastName
from sale
group by CustomerID
limit 10
;
```

```python
%%sql
select sum(SalePrice*Quantity) as Revenue,
       CustomerID,
       FirstName,
       LastName
from sale
group by CustomerID
order by Revenue desc
limit 10
;
```

```python
%%sql
select sum(SalePrice*Quantity) as Revenue,
       CustomerID,
       FirstName,
       LastName
from sale
group by CustomerID
having sum(SalePrice*Quantity)>2000
order by Revenue desc
limit 10
;
```
