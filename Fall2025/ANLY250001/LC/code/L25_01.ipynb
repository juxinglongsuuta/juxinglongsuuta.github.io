```python
# 5
%%sql
select count(*) as N,
       State,
       max(Population) as MAX,
       min(Population) as MIN,
       sum(Population) as Total,
       avg(Population) as AVG
from data
where State='Texas'
;
```

```python
%%sql
select count(*) as N,
       State,
       max(Population) as MAX,
       min(Population) as MIN,
       sum(Population) as Total,
       avg(Population) as AVG
from data
group by State
;
```

```python
%%sql
select max(Population) as MAX,
       City,
       State
from data
group by State
order by MAX desc
;
```
