```python
# for loop to collect a few grades, then calculate the average grade
# define accumulator/counter variables
sumG=0.0 # sum of the grades
nStu=0

print("Input the number of students:")
nStr=input()
nStu=int(nStr)


# for loop to collect a few grade from keyboard
for i in range(nStu):
  print(f"\nInput student {i+1} grade:")
  gStr=input()
  g=float(gStr)

  # deposit into a bank account
  sumG=sumG+g

# calculate the average
a = sumG/nStu
print(f"\nNumber of students: {nStu}")
print(f"Average grade: {a:.2f}")
```

```python
# student grades

# define variables: counter/accumulator variables
N=0 # number of students
nA=0 # number of A's
nB=0
nC=0
nD=0
nF=0

sumG=0 # sum of all grades
sumGA=0 # sum of all A grades
sumGB=0
sumGC=0
sumGD=0
sumGF=0

# while loop
while True:
  # collect one more grade
  N=N+1 #
  print(f"\nInput student {N} grade:")
  # process
  gstr=input()
  g=float(gstr)
  sumG=sumG+g #
  if g>=90:
    print("Letter grade: A")
    nA=nA+1 #
    sumGA=sumGA+g #
  elif g>=80:
    print("B")
    nB=nB+1
    sumGB=sumGB+g
  elif g>=70:
    print("C")
    nC=nC+1
    sumGC=sumGC+g
  elif g>=60:
    print("D")
    nD=nD+1
    sumGD=sumGD+g
  elif g<60: # else:
    print('F')
    nF=nF+1
    sumGF=sumGF+g

  print("\nAnother student's grade? (y/n)")
  userChoice=input()# string
  if userChoice=="n":
    break# stop the loop


# summary information
print("\n\nSummary Information")
print(f"Number of students: {N}")

a=sumG/N
print(f"Average grade: {a:.2f}")
a=sumGA/nA
print(f"Average A grade: {a:.2f}")
a=sumGB/nB
print(f"Average B grade: {a:.2f}")
a=sumGC/nC
print(f"Average C grade: {a:.2f}")
a=sumGD/nD
print(f"Average D grade: {a:.2f}")
a=sumGF/nF
print(f"Average F grade: {a:.2f}")

# ratio
r=nA/N
print(f"\n A: {nA}/{N}, {r:.2%}")
r=nB/N
print(f"B: {nB}/{N}, {r:.2%}")
r=nC/N
print(f"C: {nC}/{N}, {r:.2%}")
r=nD/N
print(f"D: {nD}/{N}, {r:.2%}")
r=nF/N
print(f"F: {nF}/{N}, {r:.2%}")
```
