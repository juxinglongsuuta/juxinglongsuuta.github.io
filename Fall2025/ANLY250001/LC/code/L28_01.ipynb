```python
%%sql
select *
from sale
limit 10
;
```

```python
## top 10 manufacturer with largest revenue
%%sql
select ManufacturerName,
       sum(SalePrice*Quantity) as TotalRevenue,
       sum(Quantity) as TotalPairs
from sale
group by ManufacturerName
order by TotalRevenue desc
limit 10
;
```

```python
%%sql
select ManufacturerName,
       sum(SalePrice*Quantity) as TotalRevenue,
       sum(Quantity) as TotalPairs
from sale
group by ManufacturerName
order by TotalPairs desc
limit 10
;
```

```python
# 6
%%sql
select ProductName,
       max(SalePrice) as SalePrice
from sale
;
```

```python
# 7
%%sql
select sum(SalePrice*Quantity) as TotalRevenue,
       ManufacturerName
from sale
group by ManufacturerName
;
```

```python
%%sql
select sum(SalePrice*Quantity) as TotalRevenue,
       ManufacturerName
from sale
group by ManufacturerName
order by TotalRevenue desc
;
```

```python
%%sql
select sum(SalePrice*Quantity) as TotalRevenue,
       ManufacturerName
from sale
group by ManufacturerName
order by sum(SalePrice*Quantity) desc
limit 3
;
```

```python
# 8
#
%%sql
select sum(Quantity) as TotalPairs,
       CustomerID,
       FirstName,
       LastName
from sale
group by CustomerID
order by TotalPairs desc
limit 1
;
```

```python
# 9
# having where
%%sql
select sum(Quantity) as TotalPairsSold,
       OrderID
from sale
group by OrderID
having sum(Quantity)>5
order by TotalPairsSold desc
;
```

```python
# 10
%%sql
select ProductName,
       Color,
       SalePrice
from sale
where Color='Green' and
      SalePrice<20
order by SalePrice desc
limit 10
;
```

```python
# 11
%%sql
select ProductName,
       SalePrice
from sale
where SalePrice>=100 and
      SalePrice<=150
order by SalePrice desc
limit 10
;
```
