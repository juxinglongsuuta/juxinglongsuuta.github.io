```python
%%sql
-- max min count sum avg
select count(*) as N,
       max(Population) as MAX,
       min(Population) as MIN
from data
;
```

```python
%%sql
-- max min count sum avg
-- group by: divide a signle table into multiple tables
select count(*) as N,
       max(Population) as MAX,
       min(Population) as MIN,
       State
from data
group by State
;
```

```python
%%sql
-- max min count sum avg
-- group by: divide a signle table into multiple tables
-- asc desc
select count(*) as N,
       max(Population) as MAX,
       min(Population) as MIN,
       State
from data
group by State
order by N desc
;
```

```python
%%sql
-- max min count sum avg
-- group by: divide a signle table into multiple tables
-- asc desc
select count(*) as N,
       max(Population) as MAX,
       min(Population) as MIN,
       State
from data
group by State
having count(*)>30--where
order by N desc
;
```
