```python
states=["Utah","Nevada",'Texas']
```

```python
r=df1.groupby(by=["State"])#
for i in states:# i="Utah" i="Nevada" i="Texas"
  print("\nOne iteration")
  print(f"The current state: {i}")
  sdf=r.get_group(i)#
  sdf.to_csv(f"d1/{i}.csv",index=False)
  print(sdf)
```

```python
r=df1.groupby(by=["State"])#
states=r.groups.keys()
for i in states:# i="Utah" i="Nevada" i="Texas"
  print("\nOne iteration")
  print(f"The current state: {i}")
  sdf=r.get_group(i)#
  sdf=sdf.sort_values(by=["Population"],ascending=[False])
  #
  sdf.to_csv(f"d1/{i}.csv",index=False)
  print(sdf.head(3))
```
