```python
# while loop to collect a few grades
# then calculate the average grade
# define variables
nStu=0 # number of stduents
sumG=0 # sum of the grades

while True: # infinite loop
  # loop body
  nStu=nStu+1
  print(f"\nInput student {nStu} grade:")
  gStr=input()
  g=float(gStr)

  sumG=sumG+g

  # another student grade
  print(f"\nAnother student grade? (y/n)")
  userChoice=input()
  if userChoice=='n':
    break # stop the loop


a = sumG/nStu
print(f"\nNumber of students: {nStu}")
print(f"Average grade: {a:.2f}")

```

```python
# sales analysis
# cog: 60
# gear: 200

# define variables (counter/accumulator variables)
N=0 # number of orders
nC=0 # total number of cogs sold
nG=0 # total number of gears sold

nL1=0 # number of orders subtotal < 1000
nL2=0 # 1000-2000
nL3=0 # > 2000

sumS=0 # sum of subtotals
sumC=0 # sum of revenue for cogs
sumG=0 # sum of revenue for gears

# collect orders
while True:
  # one more order
  N=N+1 #
  print(f"\nNumber cogs in Order {N}:")
  cstr=input()
  c=int(cstr)
  nC=nC+c
  print(f"Number gears in Order {N}:")
  gstr=input()
  g=int(gstr)
  nG=nG+g

  s=60*c+200*g
  sumS=sumS+s #
  sumC=sumC+60*c #
  sumG=sumG+200*g #

  if s<1000:
    print(f"Subtotal: ${s:.2f}, below $1000")
    nL1=nL1+1 #
  elif s<=2000:
    print(f"Subtotal: ${s:.2f}, between $1000-$2000")
    nL2=nL2+1
  elif s>2000:#else:
    print(f"Subtotal: ${s:.2f}, above $2000")
    nL3=nL3+1

  print(f"\n\nAnother order? (y/n)")
  userChoice=input()
  if userChoice=='n':
    break# stop the loop

# summary information
print("\n\nSummary Information")
print(f"N orders: {N}")
print(f"N cogs: {nC}")
print(f"N gears: {nG}")

print(f"\nSum subtotal: ${sumS:.2f}")
print(f"Sum revenue for cogs: ${sumC}")
print(f"Sum revenue for gears: ${sumG}")
print(f"Average subtotal: ${sumS/N:.2f}")

print(f"\nL1: {nL1}/{N}, {nL1/N:.2%}")
print(f"L2: {nL2}/{N}, {nL2/N:.2%}")
print(f"L3: {nL3}/{N}, {nL3/N:.2%}")
```
