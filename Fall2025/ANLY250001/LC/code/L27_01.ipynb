```python
# 4
%%sql
select sum(SalePrice*Quantity) as TotalRevenue
from sale
;
```

```python
%%sql
select sum(SalePrice*Quantity) as TotalRevenue,
       Category
from sale
group by Category
having sum(SalePrice*Quantity)>10000
order by TotalRevenue desc
;
```

```python
# where: NO aggregation funtion as condition
# having: aggregation function as condition in having
%%sql
select sum(SalePrice*Quantity) as TotalRevenue,
       Category
from sale
group by Category
having sum(SalePrice*Quantity)>10000
order by sum(SalePrice*Quantity) desc
;
```

```python
# 5
%%sql
select sum(Quantity) as NPairsSold,
       Category
from sale
group by Category
;
```

```python
%%sql
select sum(Quantity) as NPairsSold,
       Category
from sale
group by Category
order by NPairsSold desc
;
```

```python
%%sql
select sum(Quantity) as NPairsSold,
       Category
from sale
group by Category
having sum(Quantity)>1000
order by NPairsSold desc
;
```
